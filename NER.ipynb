```python
import spacy
from spacy import displacy
```

```python
from collections import Counter
nlp=spacy.load('en_core_web_sm')
```

```python
raw_text="The Indian Space Research Organisation or is the national space agency of India, headquartered in Bengaluru. It operates under Department of Space which is directly overseen by the Prime Minister of India while Chairman of ISRO acts as executive of DOS as well."
```

```python
text1=nlp(raw_text)
text1
```

```python
for word in text1.ents:
    print(word.text,word.label_)
```

```python
spacy.explain("ORG")
```

```python
spacy.explain("GPE")
```

```python
displacy.render(text1,style="ent",jupyter=True)
```

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
def url_to_string(url):
    res=requests.get(url).text
    soup=BeautifulSoup(res,'lxml')
    for script in soup(["script","style",'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+',soup.get_text()))
```

```python
ny_bb = url_to_string("https://www.zeebiz.com/markets/currency/news-cryptocurrency-news-today-june-12-bitcoin-dogecoin-shiba-inu-and-other-top-coins-prices-and-all-latest-updates-158490")
article = nlp(ny_bb)
len(article.ents)
```

```python
ny_bb
```

```python
for word in article.ents:
    print(word.text,word.label_)
```

```python
spacy.explain("PERSON")
```

```python
spacy.explain("CARDINAL")
```

```python
displacy.render(article,style="ent",jupyter=True)
```
